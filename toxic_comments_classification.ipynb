```python
# Removing stopwords

comment_text = train.comment_text.tolist()

def remove_stop_words(text):
    stopwords_spacy = spacy.lang.en.stop_words.STOP_WORDS
    doc = nlp(text)
    filtered_tokens = []
    for token in doc:
        if token.is_stop:
            continue
        filtered_tokens.append(token.text)
    return " ".join(filtered_tokens)

comment_text_without_stopwords = list()

for i in comment_text:
    comment_text_without_stopwords.append(remove_stop_words(i))

comment_text_without_stopwords
```

```python
train['comment_text_without_stopwords'] = comment_text_without_stopwords
train.head()
```
